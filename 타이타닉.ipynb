## 데이터 스케일링
### 모델 학습 전 피처의 스케일을 맞추는 것이 중요

1.   **왜 데이터 스케일링이 필요한가?**

  많은 머신러닝 모델, 특히 **선형 모델**(예: 로지스틱 회귀, 선형 회귀), **신경망 모델**(예: 텐서플로우로 구축한 딥러닝 모델) 에서는 입력 데이터(피처)의 **스케일(크기차이)**가 중요.
  
  그 이유는 모델이 데이터의 크기에 따라 성능이 달라질 수 있기 때문.

  - 예를 들어, **나이(age)**와 **요금(fare)** 같은 피처들이 있다고 할때, 나이는 보통 0~100 사이의 값을 가지는데, 요금은 수백, 수천 심지어 수만 원에 이를 수 있음. 만약 스케일을 맞추기 않으면, 모델이 요금 같은 큰 숫자에 더 큰 가중치를 두고, 나이에 대해서는 상대적으로 덜 중요한 특성으로 간주할 수 있음.

  - **스케일 차이가 큰 피처들**이 존재하면, 학습 과정에서 모델이 특정 피처에 대해 과도하게 또는 부족하게 반응 할 수 있음. 그 결과, **모델 성능**이 떨어지거나 학습이 제대로 되지 않을 수 있음.
2.   **스케일링이랑 무엇인가?**

  **스케일링**은 피처들의 값이 서로 다른 범위에 있을 때, 그 범위를 비슷한 범위로 변환하는 작업을 말함. 보통 두 가지 방법이 많이 사용됨

  **(1) 표준화(Standardization)**

    표준화는 각 피처를 **평균 0, 표준편차 1**로 변환하는 방법임. 즉, 각 피처의 값에서 평균을 빼고, 표준편차로 나누는 것.
    데이터를 0을 중심으로, 분포가 1인 표준 정규 분포하는것.
    이상치의 영향을 덜 받음.

    ex)
      - 데이터가 정규 분포(가우시안 분포)를 따를 때
      - 선형 모델(로지스틱 회귀, 선형 회귀) 또는 PCA(주성분 분석)를 사용할 때
      - 특성의 단위(scale)가 다를 때 (예: 키(cm)와 몸무게(kg)처럼 크기가 다르면 스케일링 필요)

    사용 사례)
      - PCA를 적용하여 차원 축소할 때
      - 선형 회귀, 로지스틱 회귀 등 가중치(weight)를 기반으로 하는 모델
      - K-Means 같은 클러스터링 알고리즘
    
    ```
    import numpy as np
    from sklearn.preprocessing import StandardScaler

    # 예시 데이터
    data = np.array([[1, 2], [3, 4], [5, 6]])

    # StandardScaler 객체 생성
    scaler = StandardScaler()

    # 표준화
    scaled_data = scaler.fit_transform(data)

    print(scaled_data)
    ---
    [[-1.22474487 -1.22474487]
    [ 0.          0.        ]
    [ 1.22474487  1.22474487]]
    ```

  **(2) 정규화(Normalization)**
  
    정규화는 각 피처를 0과 1사이로 변환하는 방법.
    최소값과 최대값을 기준으로 비율을 계산.

    ex)
      - 값의 범위가 일정하지 않은 경우
      - 데이터가 특정 범위(예: [0, 1] 또는 [-1, 1])에 있어야 하는 경우
       (예: 신경망 모델에서는 입력값이 0~1 사이에 있으면 학습이 더 안정적)

    사용 사례)
      - 이미지 픽셀 값을 [0,1]로 변환
      - 신경망(NN) 모델 학습 데이터 전처리
      - KNN, SVM 등 거리 기반 알고리즘 사용 시
    
    ```
    import numpy as np
    from sklearn.preprocessing import MinMaxScaler

    data = np.array([[1, 2], [3, 4], [5, 6]])
    
    # MinMaxScaler 객체 생성
    scaler = MinMaxScaler()
    
    #정규화
    scaled_data = scaler.fit_transform(data)
    
    print(scaled_data)
    ---
    [[0.  0. ]
    [0.5 0.5]
    [1.  1. ]]
    ```


3.   **스케일링의 이점**

  - 모델 성능 개선: 스케일링을 하면 모델이 모든 피처를 동일한 중요도로 다룰 수 있음. 특히, 신경망에서는 가중치 업데이트가 균형 있게 이루어지기 떄문에 훈련이 더 빨라지고, 성능도 좋아질 수 있음.

  - 학습 안정성 향상: 피처들이 비슷한 범위로 변환되면, 경사하강법(gradient descent) 같은 최적화 알고리즘이 더 안정적으로 작동함. 이로 인해 모델이 더 빠르고 안정적으로 학습할 수 있음.

  - 모델 최적화: 특히 **신경망**이나 **K-최근접 이웃(k-nn)**모델 같은 경우, 거리 기반 계산이 많기 때문에, 피처들의 스케일이 다르면 결과가 왜곡될 수 있음. 이때 스케일링을 통해 모델의 성능을 향상시킬 수 있음.

4.   **스케일링을 하지 않으면?**

  스케일링을 하지 않으면, 모델이 **스케일이 큰 피처**에 더 많은 가중치를 두고 학습할 수 있음. 예를 들어, `fare`와 같은 큰 숫자를 가지는 피처가 모델의 예측에 더 영향을 미칠 수 있고, `age` 같은 작은 숫자 피처는 상대적으로 무시될 가능 성이 있음.

  또한, 신경망에서 학습 속도가 느려지거나 수렴하지 않는 문제가 발생할 수 있음.

----

따라서, 스케일링을 통해 피처들이 동일한 범위로 맞춰지면, 모델이 더 **효율적으로, 빠르게, 정확하게** 학습할 수 있음.
특히 **신경망**에서는 학습 성능에 아주 중요한 영향을 미친다고 할 수 있음.

````python
# 데이터 스케일링
# 모델 학습 전에 피처의 스케일을 맞추는 것이 중요 StandardScaler를 사용해서 피처를 표준화

# 스케일러 객체 생성
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

## `fit_transform()` vs `transform()` 차이
### `fit_transform()`

```
X_train = scaler.fit_transform(X_train)
```

- `fit()`: `X_train` 데이터를 사용해 **평균(mean)**과 **표준편차(std)**를 계산
- `transform()`: 계산된 평균과 표준편차를 사용해 `X_train`을 표준화


즉, X_train에서만 평균과 표준편차를 계산하고 적용

### `transform()`

```
X_test = scaler.transform(X_test)
```

- `X_train`에서 구한 **평균과 표준편차를 그대로 사용**해서 `X_test`를 변환


즉, 테스트 데이터는 새로운 평균과 표준편차를 계산하면 안 되고, 학습 데이터(`X_train`)에서 구한 값으로 변환


### 만약 `X_test`에도 `fit_transform()`을 하면?

```
X_test = scaler.fit_transform(X_test)  # ❌ 이렇게 하면 안 됨!

```

이렇게 하면 **테스트 데이터의 평균과 표준편차를 따로 계산해서 변환**하기 때문에
**훈련 데이터와 테스트 데이터의 스케일이 다르게 변환될 수 있음**


그럼 모델이 일반화되지 못하고 과적합(overfitting)될 가능성이 커짐





## TensorFlow 모델 구축

### 1️⃣ `tf.keras.Sequential([...])` 사용

``` python
model = tf.keras.Sequential([...])

```

- `Sequential`은 **"순차적인"**모델을 만들 때 사용
- 즉, 레이어(layer)를 한 줄로 쌓아 올리는 방식
- **입력층 -> 중간층 -> 출력층** 이런식으로 순차적으로 진행될 때 적합
- 복잡한 네트워크 구조(CNN, RNN등)는 `Sequentail`보다는 `Functional API`를 쓰기도 함

### 2️⃣ layer의 Dense

``` python
tf.keras.layers.Dense(16, activation='relu', input_shape=(X_train.shape[1],))
tf.keras.layers.Dense(8, activation='relu')
tf.keras.layers.Dense(1, activation='sigmoid')

```

- `Dense()`는 **완전 연결층(Fully Connected Layer, FC Layer)**을 의미
- 각 뉴련이 **이전 층의 모든 뉴런과 연결**되어 있음
- 여기서 중요한 건 **뉴런 갯수와 활성화 함수(activation function)**임

### 🔹 (1) 입력층

```python
tf.keras.layers.Dense(16, activation='relu', input_shape=(X_train.shape[1],))

```

- 뉴런 갯수: 16개 (적절한 피처 수를 고려한 적당한 크기)
- 활성화 함수: ReLU (`relu`)
  - `ReLU` 는 음수를 0으로 바꾸고 양수는 그대로 유지해서 **기울기 소실 문제**를 방지
- `input_shape=(X_train.shape[1],)`:
  - 입력 데이터의 **특성(feature) 갯수**만큼 입력을 받음
  - 예를 들어, `X_train.shape[1] == 9`라면 입력 뉴런이 9개라는 뜻

---
### 🔹 (2) 중간층(Hidden Layer)

```python
tf.keras.layers.Dense(8, activation='relu')

```

- 뉴런 갯수: 8개 (앞 층보다 줄여서 압축된 표현을 학습)
- 활성화 함수: ReLU (`relu`)
- 중간층을 추가하는 이유?
  - 데이터의 복잡한 패턴을 학습하기 위해 사용
  - 하나의 `Dense()`층만 있으면 너무 단순한 모델이 될 수 있음
  - 층을 늘리면서 학습할 수 있는 표현력이 증가

---
### 🔹 (3) 출력층(Output Layer)

```python
tf.keras.layers.Dense(1, activation='sigmoid')

```

- 뉴런 갯수: 1개 (이진 분류이기 때문에 0 또는 1을 출력해야 함)
- 활성화 함수: Sigmoid (`sigmoid`)
  - `sigmoid`는 **출력값을 0~1 사이로 변환**하여 확률처럼 사용 가능
  - 예: `0.9` -> 90% 확률로 생존, `0.2` -> 20% 확률로 생존
  - `sigmoid`를 쓰면 `0.5`기준으로 `0`(사망)/`1`(생존)으로 분류할 수 있음

---

### 3️⃣ `model.compile()`이란

``` python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

```

이 단계는 **모델을 학습할 준비**를 하는 과정

세 가지 중요한 요소가 있음.


### 🔹 (1) 옵티마이저(`optimizer='adam'`)

- Adam(Adaptive Moment Estimation) 사용
- 학습률을 자동으로 조절하여 효율적으로 최적화해줌
- 대부분의 경우 `adam`이 기본적으로 좋은 성능을 보임

👉 **가중치를 업데이트하는 방식**을 결정


---

### 🔹 (2) 손실 함수 (`loss='binary_crossentropy'`)

- `binary_crossentropy`: 이진 분류(Binary Classification)에 최적화된 손실 함수
- 수식적으로 `-y log(p) - (1 - y) log(1 - p)`형태로, 정답 확률이 높을수록 손실이 작아짐

👉 **모델이 얼마나 틀렸는지 평가**하는 기준


---


### 🔹 (3) 평가 지표 (`metrics=['accuracy']`)

- `accuracy`: **정확도(Accuracy)**를 평가 지표로 사용
- 예측값(`0` 또는 `1`)과 실제값이 얼마나 일치하는지 측정

👉 **모델이 얼마나 잘 학습하고 있는지 판단**하는 기준


---

### 4️⃣ `model.fil()`에서 `epochs=50, batch_size=16`설정

``` python
model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

### 🔹 (1) `epochs=50`

- `epoch`는 **전체 데이터를 몇 번 반복해서 학습할지**를 정하는 것
- `epochs=50`이면 **모든 훈련 데이터를 50번씩 반복해서 학습**한다는 뜻
- 너무 적으면 학습이 부족할 수 있고, 너무 많으면 과적합(overfittng) 가능성이 있음
- 보통 50~100 정도에서 조절하면서 쵲거의 값을 찾음

👉 모델이 충분히 학습 할 수 있도록 반복 횟수를 조절


---

### 🔹 (2) `batch_size=16`

- `batch_size`는 **한 번에 몇 개의 샘플을 학습할지** 정하는 값
- `batch_size=16`이면 **16개 데이터씩 묶어서 학습**
- 너무 작으면 훈련이 느려지고, 너무 크면 메모리 사용량이 커짐
- 일반적으로 `16`, `32`, `64` 같은 값을 사용

👉 적절한 크기로 데이터를 나눠서 학습을 진행

## 📌 모델 구성 정리  

| 개념 | 설명 |
|:------|:------|
| `Sequential()` | 순차적으로 층을 쌓는 모델 |
| `Dense(16, activation='relu', input_shape=(X_train.shape[1],))` | 입력층 (16개 뉴런, `ReLU` 활성화 함수) |
| `Dense(8, activation='relu')` | 중간층 (8개 뉴런, `ReLU`) |
| `Dense(1, activation='sigmoid')` | 출력층 (1개 뉴런, `sigmoid` 활성화 함수) |
| `optimizer='adam'` | `Adam` 옵티마이저 사용 |
| `loss='binary_crossentropy'` | 이진 분류 손실 함수 |
| `metrics=['accuracy']` | 정확도를 평가 지표로 사용 |
| `epochs=50` | 전체 데이터셋을 50번 학습 |
| `batch_size=16` | 16개씩 묶어서 학습 |




````python
X_train.shape
````

````python
# TensorFlow 모델 구축

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),  # 입력 정의
    tf.keras.layers.Dense(16, activation='relu'), # 첫 번째 Dense층
    tf.keras.layers.Dense(8, activation='relu'), # 중간층
    tf.keras.layers.Dense(1, activation='sigmoid') # 출력층(이진 분류)
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 모델 학습
model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
````

````python
# 평가
loss, accuracy = model.evaluate(X_test,y_test)
print(f"테스트 정확도: {accuracy:.4f}")
print('loss: ', loss)

# 예측 수행
y_pred = (model.predict(X_test) > 0.5).astype(int)

# 결과 확인
print(y_pred[:10]) ## 예측된 10개의 생존 여부 출력
print('------')
print(y_test[:10])
````

위 블로그에서 동일하게 keras.Sequential을 사용하는데
모델 작성시 두가지를 더 추가해서 학습시켜서 진행한다.

Dropout과 Batchnormalization이다.


해당 블로거에서 말하기로는

- Dropout은 Overfitting,model combination 문제를 해결하기 위해 등장한 개념이고 신경망의 뉴런을 랜덤하게 부분적으로 생략시킨다고 한다.
- 신경망을 망각시키는 느낌
- Batchnormalization은 배치 정규화인데, 활성화 함수의 활성화 값이나 출력 값을 정규분포로 만들어줘서 noise를 추가하는 개념이란다. (이미 스케일링할때 표준화 작업을 했는데 이게 또 필요한건가...? )
- 학습을 할 때마다 활성화 값이나 출력 값을 정규화 하므로, 초기화(가중치 초깃값) 문제의 영향을 덜 받게 해준다.
- 학습률(Learning Rate)을 높게 설정할 수 있으므로, 학습 속도가 개선된다.
- Overfiggin 위험을 줄일 수 있다.
- 가중치 소실(Gradient Vanishing)문제를 해결해준다.
